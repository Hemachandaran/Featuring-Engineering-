### LABEL ENCODING 

```python
l = ['Switch to order menu','Is delivering now','Has Online delivery','Has Table booking']
for i in l:
    df[i] = df[i].map({'Yes': 1, 'No': 0})
df.sample(6)
```

### ORDINAL ENCODING 

```python
encoder_RT = OrdinalEncoder(categories=[['Not rated','Poor','Average' ,'Good' ,'Very Good'  ,'Excellent']])
encoder_RC = OrdinalEncoder(categories=[['Red','Orange', 'Yellow' ,'White', 'Green' ,'Dark Green']])

df["encoded_Rating_color"] = encoder_RC.fit_transform(df[["Rating color"]])
df["encoded_rating_text"] = encoder_RT.fit_transform(df[["Rating text"]])

df = df.drop(["Rating color","Rating text"],axis=1)

df.sample(6)
```

```python
df.describe(include="object")
```

```python
t_encoder = TargetEncoder()
df_te1 = t_encoder.fit_transform(df[['City','Address','Locality','Locality Verbose']],df['Aggregate rating'])
encode_data_2 = pd.DataFrame(df_te1,columns= ['City_encoded','Address_encoded','Locality_encoded','Locality Verbose encoded'])
df = pd.concat([df,encode_data_2],axis=1)
# Calculate the mean target for each restaurant
target_encoded = df.groupby('Restaurant Name')['Aggregate rating'].mean().to_dict()
Currency_encoded = df.groupby('Currency')['Aggregate rating'].mean().to_dict()
df['Encoded_Res_name'] = df['Restaurant Name'].map(target_encoded)
df['Encoded_Currency'] = df['Currency'].map(Currency_encoded)
df= df.drop(['City','Address','Locality','Locality Verbose',"Restaurant Name","Currency"],axis=1)
```

```python
df.sample(6)
```

### MultiLabelBinarizer

```python
df['Cuisine_split']=df['Cuisines'].str.split(', ')
mlb = MultiLabelBinarizer()
df_02 = mlb.fit_transform(df['Cuisine_split'])
dg = pd.DataFrame(df_02, columns=mlb.classes_)
# joining the entire falt feature in to single feature 
# encoding the feature in to binary value
dg
```

```python
df = pd.concat([df, dg],axis=1)
```
